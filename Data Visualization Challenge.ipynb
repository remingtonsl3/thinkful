# Visualization of Craft Canned Beer Data

I have chosen to look at data from January 2017 from CraftCans.com. I found this data on kaggle.com.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

### I created a dataframe from the values in the beers.csv files

```python
beers_df = pd.read_csv('beers.csv')
```

### I then wanted to see what the sizes of cans were in the craft beer world

Unsurprisingly, since the standard beer can in the US 12 oz., with 16 oz. being common as well, our data show that craft beers tend to fall in that range, with a couple monster cans out there in the >20-ounce range.

However, it was interesting that the median appeared to be equal to the 25% quartile. I took a look at the counts with a histogram to see if it that made sense. As we see, the large percentage of 12 oz cans shows that the low median makes. A single 8.4 oz. value explains the lower whisker.

```python
plt.figure(figsize=(10, 5))
           
plt.subplot(121)
plt.boxplot(beers_df['ounces'])
plt.ylabel('Ounces')
plt.title('Ounces Per Can of Craft Beer')

plt.subplot(122)
plt.hist(beers_df['ounces'], bins=30)
plt.xlabel('Ounces')
plt.title('Number of Cans of Different Volumes')
plt.show()
```

# Most common styles of beer

I thought I would take a look at the count of the different types of beer. I looked up how to use 'value_counts' to make a bar chart. It was quickly obvious that this was a useless chart because there were way to many styles to make it readable, so I tried to figure out how to just do a top ten. It took a bit of searching, but the 'nlargest' method solved my problem.

It seems that the bitter beer fans are in good shape right now. The American IPA and other Pale Ales seem to have a lock on the craft beer world right now.

```python

beers_df['style'].value_counts().nlargest(10).plot(kind = 'bar')
plt.xlabel('Style')
plt.ylabel('Count')
plt.title('Beer Style Count')
plt.show()
```
