```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Chisquare Distribution

```python
chivar = np.random.chisquare(7, 100)

plt.hist(chivar)

# Add a vertical line at the mean.
plt.axvline(chivar.mean(), color='red', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(chivar.mean() + chivar.std(), color='red', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(chivar.mean() - chivar.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

## Part 2

1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.

2. Add them together to create a third variable.

3. Graph the third variable using a histogram.

4. Compute the mean and standard deviation and plot them as vertical lines on the histogram.

5. Evaluate the descriptive statistics against the data.


```python

# mean of 5, s of 0.5; mean of 10 s of 1
pop1 = np.random.normal(5, 0.5, 1000)
pop2 = np.random.normal(10 , 1, 1000)
pop3 = pop1 + pop2


# Add a vertical line at the mean.
plt.axvline(pop3.mean(), color='red', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(pop3.mean() + pop3.std(), color='red', linestyle='dashed', linewidth=2)

#Add a vertical line at one standard deviation below the mean.
plt.axvline(pop3.mean() - pop3.std(), color='red', linestyle='dashed', linewidth=2)

# Plot a histogram.
plt.hist(pop3)

# Print the histogram.
plt.show()
```

Data shows a normal distribution shape, but isn't exact, reflecting the random nature of the data.
